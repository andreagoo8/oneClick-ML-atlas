```python
import digitalhub as dh
project = dh.get_or_create_project('tp-ocml')
```

##  Dataset

```python
train_data = project.log_dataitem(name="data", kind='table', source='data.csv')
train_data.key
```

```python
# upload = project.log_artifact(name='big-data', kind='artifact', source="tiles")
```

## Train

### Register Train Function


```python
func_train = project.new_function(
    name="train",
    kind="python", 
    python_version="PYTHON3_10",
    requirements="requirements.txt",
    code_src="git+https://github.com/andreagoo8/oneClick-ML-atlas#digitalhub_integration",
    handler= "train:train"
)
```

```python
build_run = func_train.run(action="build")
```

```python
#func_train=project.get_function('train')
```

### Execute

```python
train_run = func_train.run(action="job",
                     inputs={"di": train_data.key},
                     # profile="1xA5000",
                     parameters={
                         "target_model_name": "first_integrated_model",
                         "n_iter": 1,
                         "classifiers_list": "logisticregression",
                         "shap_analysis": False, 
                         "n_of_features":8
                     },
                     resources={"mem": "6Gi"},
                     # volumes=[{
                     #     "volume_type": "persistent_volume_claim",
                     #     "name": "train-volume",
                     #     "mount_path": "local-data",
                     #     "spec": {
                     #         "size": "10Gi"
                     #     }}],
                     local_execution=False)
```
